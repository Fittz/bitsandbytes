```python
import random
data = [random.randint(1,10000) for x in xrange(10000)]
```

```python
index = DatetimeIndex(start='01-01-2003', periods=len(data), freq='T')
```

```python
s = Series(data, index=index)
```

```python
s
```

```python
s.std()
```

```python
s_daily = s.resample('D')
```

```python
s_daily
```

```python
daily_sum = s.resample('D',how='sum')
```

```python
daily_sum
```
